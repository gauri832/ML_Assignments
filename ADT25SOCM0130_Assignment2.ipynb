## ðŸ”¹ Step 2: Load Dataset from UCI

```python
df = pd.read_csv("winequality-red.csv", sep=';')
df = df.rename(columns={"quality": "class"})
df.head()
```

## ðŸ”¹ Step 7: Remove Duplicate Records

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print('Duplicates removed:', before - after)
```

## ðŸ”¹ Step 8: Encode Categorical Column

```python
encoder = LabelEncoder()
df['class'] = encoder.fit_transform(df['class'])
df.head()
```

## ðŸ”¹ Step 9: Feature Scaling

```python
scaler = StandardScaler()
# Select all columns except the 'class' column for scaling
feature_cols = df.columns.drop('class')
df[feature_cols] = scaler.fit_transform(df[feature_cols])
df.head()
```
